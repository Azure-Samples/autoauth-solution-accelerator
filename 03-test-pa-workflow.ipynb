```python
import concurrent.futures
import asyncio

# Initialize counters
success_count = 0
failure_count = 0

# List to store results
results = []

# Function to run the pipeline
async def run_pipeline(iteration):
    try:
        print(f"Running iteration {iteration}...")
        await pa_pipeline.run(uploaded_files=files, use_o1=True)
        return f"Iteration {iteration}: Success"
    except Exception as e:
        return f"Iteration {iteration}: Failure - {str(e)}"

# Wrapper to run the async function in a thread
def run_pipeline_sync(iteration):
    return asyncio.run(run_pipeline(iteration))

# Run the function 20 times with 4 parallel executions
with concurrent.futures.ThreadPoolExecutor(max_workers=4) as executor:
    futures = [executor.submit(run_pipeline_sync, i+1) for i in range(20)]
    for future in concurrent.futures.as_completed(futures):
        result = future.result()
        results.append(result)
        if "Success" in result:
            success_count += 1
        else:
            failure_count += 1
        print(result)

# Log the summary
print(f"Total Successes: {success_count}")
print(f"Total Failures: {failure_count}")
```
