```python
from src.documentintelligence.document_intelligence_helper import (
    AzureDocumentIntelligenceManager,
)
import os
from dotenv import load_dotenv

dotenv_path = "../../.azure/.hls-autoauth/.env"
load_dotenv(dotenv_path)
```

```python
document_intelligence_client = AzureDocumentIntelligenceManager(
    azure_endpoint=os.getenv("AZURE_DOCUMENT_INTELLIGENCE_ENDPOINT"),
    azure_key=os.getenv("AZURE_DOCUMENT_INTELLIGENCE_KEY"),
    storage_account_name=os.getenv("AZURE_STORAGE_ACCOUNT_NAME"),
    container_name="pre-auth-policies",
    account_key=os.getenv("AZURE_STORAGE_ACCOUNT_KEY"),
)
```

```python
folder_path = "../../utils/data/cases/policies"  # Change this to your actual folder path
pdf_binary_map = {}

for i in range(1, 6):  # Loop from 001.pdf to 005.pdf
    file_name = f"{i:03}.pdf"  # Format as 001.pdf, 002.pdf, etc.
    file_path = os.path.join(folder_path, file_name)

    # Read the binary content of the PDF file if it exists
    if os.path.exists(file_path):
        with open(file_path, "rb") as pdf_file:
            pdf_binary_map[file_name.split('.')[0]] = pdf_file.read()
```

```python
pdf_policy_text_map = {}
for document, blob_content in pdf_binary_map.items():
    policy_text = document_intelligence_client.analyze_document(
        document_input=blob_content,
        model_type="prebuilt-layout",
        output_format="markdown",
    )
    pdf_policy_text_map[document] = policy_text.content
```
